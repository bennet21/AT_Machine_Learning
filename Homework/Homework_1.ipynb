```python
import numpy as np
import matplotlib.pyplot as plt
import torch
# from torch.utils.data import TensorDataset, random_split, DataLoader
```

```python
n_samples=5000
x_values=np.random.uniform(low=0,high=2*np.pi,size=n_samples)
sin_wave=np.sin(x_values)+(0.1*np.random.randn(x_values.shape[0]))
plt.plot(x_values,sin_wave,'.')
plt.show()
```

```python
# Convert the NumPy arrays to PyTorch tensors
x_values_tensor = torch.from_numpy(x_values)
sin_wave_tensor = torch.from_numpy(sin_wave)

# Create a TensorDataset from the tensors
dataset = TensorDataset(x_values_tensor, sin_wave_tensor)

# Split the dataset
train_data, val_data, test_data = random_split(dataset, [0.5, 0.25, 0.25])
```

```python
# Creating the scatter plot
size = 2
plt.scatter(train_data[:][0], train_data[:][1], label='Training', alpha=0.5, s=size)
plt.scatter(val_data[:][0], val_data[:][1], label='Validation', alpha=0.5, s=size)
plt.scatter(test_data[:][0], test_data[:][1], label='Testing', alpha=0.5, s=size)
plt.legend()
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
